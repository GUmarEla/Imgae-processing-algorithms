### Erosion

```python
def morph_erosion(image,kernel,padding):
  hei_ima, len_ima = image.shape
  hei_ker, len_ker = kernel.shape

  ret_mat = np.zeros((hei_ima, len_ima))

  img_new = np.pad(image, pad_width=hei_ker//2, mode='constant', constant_values=0)

  # moving matrix dot
  for i in range(ret_mat.shape[0]):
    for j in range(ret_mat.shape[1]):
      ret_mat[i][j] = np.min(kernel.flatten() * img_new[i:i+hei_ker, j:j+len_ker].flatten() )


  plt.figure(figsize=(4,4))
  plt.imshow(ret_mat)
  plt.show()
```

### Dilation

```python
def morph_dilation(image,kernel,padding):
  hei_ima, len_ima = image.shape
  hei_ker, len_ker = kernel.shape

  ret_mat = np.zeros((hei_ima, len_ima))

  img_new = np.pad(image, pad_width=hei_ker//2, mode='constant', constant_values=0)

  # moving matrix dot
  for i in range(ret_mat.shape[0]):
    for j in range(ret_mat.shape[1]):
      ret_mat[i][j] = np.max(kernel.flatten() * img_new[i:i+hei_ker, j:j+len_ker].flatten() )


  plt.figure(figsize=(4,4))
  plt.imshow(ret_mat)
  plt.show()
```
