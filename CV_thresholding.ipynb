### Threshloding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("barbara.jpg")
# image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

Thresholding works for grayscae images. But if the image is colored you can change it the gray then

```python
# Convert image to gray scale
def grayscale_img(image):
  r_channel = image[:,:,0]
  g_channel = image[:,:,1]
  b_channel = image[:,:,2]

  image1 = 0.299 * r_channel + 0.587 * g_channel + 0.114 * b_channel

  return image1
```

### Convert image to gray scale

```python
# converting image to gray scae then applyinng the thresholding
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# OR
gray_image = grayscale_img(image)
```

### **Thresholding**

```python
def thresholding_image(image,threshold):

  raws, cols = image.shape[:2]
  replacement_image = np.zeros((raws, cols), dtype=np.uint8)

  if len(image.shape) == 3:
    image = grayscale_img(image)

  for i in range(raws):
    for j in range(cols):
      if image[i][j] < threshold:
        replacement_image[i][j] = 0
      else:
        replacement_image[i][j] = 255

  plt.figure(figsize=(4,4))
  plt.imshow(replacement_image, cmap='gray')
  plt.show()
```

```python
# example of colored image
thresholding_image(image,130)
```

```python
# example of gray scale image
thresholding_image(gray_image,130)
```
